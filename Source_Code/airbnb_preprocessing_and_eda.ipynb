```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset(listings)
df=pd.read_csv(r"C:\Users\dell\Downloads\listings (1).csv")
```

```python
#show first 5 rows
df.head()
```

```python
#summary statistics
df.describe()
```

```python
#basic info
df.info()
```

```python
# Remove unwanted columns

df.drop(columns='license',inplace=True)
```

```python
#checking null values

df.isna().sum()
```

```python
# removing null values

df.dropna(inplace=True)
```

```python
# Datatype conversion

# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'])

#  convert price to integer 
df['price'] = df['price'].astype('Int64')  
```

```python
# Find duplicated rows 
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
# save cleaned data to csv
df.to_csv("Airbnb.csv",index=False)
```

```python
# correlation between price and number of reviews
correlation = df['price'].corr(df['number_of_reviews'])
print(f"Correlation between price and number of reviews: {correlation:.2f}")
```

```python
# Correlation heatmap

data = ["price","number_of_reviews"]
corr = df[data].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png')  
plt.show()
```

```python
# Histogram for Distribution of Listing Prices

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=100, kde=True)
plt.title("Distribution of Listing Prices")
plt.xlabel("Price ($)")
plt.ylabel("Number of Listings")
plt.savefig('listing_price_distribution.png') #save
plt.show()
```

```python
# Distribution of Listing Availability using histogram

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='availability_365', bins=30, kde=True, color='teal')
plt.title("Distribution of Listing Availability")
plt.xlabel("Days Available")
plt.ylabel("Number of Listings")
plt.savefig('availability_distribution.png') #save
plt.show()
```

```python
# Generate Word Cloud from Review Comments

# Import necessary libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Join all non-null comments into a single string
text = " ".join(reviews['comments'].dropna().tolist())

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.savefig("wordcloud.png")  # Save it
plt.show()
```
